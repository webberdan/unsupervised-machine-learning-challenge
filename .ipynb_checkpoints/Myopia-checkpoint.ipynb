```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Pull Myopia CSV
df=pd.read_csv('myopia.csv')
df.head()
```

```python
# Remove unneeded column from dataset.
labels = df["MYOPIC"]
df.drop("MYOPIC", axis=1).head()
```

```python
# Initiate Standard Scaler
scaler = StandardScaler()
myo_scaled = scaler.fit_transform(df)
```

```python
# Utlize PCA
pca = PCA(n_components=0.90)
myoPca = pca.fit_transform(myo_scaled)

# Conver to dataframe
df_myoPca = pd.DataFrame(myoPca)


pca.explained_variance_ratio_.sum()
```

```python
# T-SNE Model
tsne = TSNE(learning_rate=25)
tsneFeatures = tsne.fit_transform(df_myoPca)

#shape t-SNE model
tsneFeatures.shape
```

```python
# plot output
plt.scatter(tsneFeatures[:,0],tsneFeatures[:,1],c=labels)
plt.show()
```
